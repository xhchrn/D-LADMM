```python
def read_logs(filename):
    use_learned = 'lskm' in filename
    use_safeguard = not ('None' in filename)
    if not use_safeguard:
        pcts = None
    with open(filename, 'r') as f:
        lines = f.readlines()
        for i, line in enumerate(lines):
            if 'values' in line or 'norms' in line:
                data = lines[i+1].strip('INFO:root:')
                data = np.array(list(map( float, data.split(',') )))
            if 'Pcts' in line:
                pcts = lines[i+1].strip('INFO:root:')
                pcts = np.array(list(map( float, pcts.split(',') )))
    return data, pcts
```
